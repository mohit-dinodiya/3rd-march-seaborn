# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular data visualization library in Python, built on top of the Matplotlib library. Some of the plots that can be created using Seaborn include:

Histograms: A histogram is a way of visualizing the distribution of a single numerical variable. Seaborn's distplot() function can be used to create histograms, which can help to identify patterns such as skewness, bimodality, and outliers.

Scatter plots: Scatter plots are used to visualize the relationship between two numerical variables. Seaborn's scatterplot() function can be used to create scatter plots, which can help to identify patterns such as correlation, outliers, and non-linearity.

Heat maps: A heat map is a way of visualizing the relationship between two categorical variables, or a categorical and a numerical variable. Seaborn's heatmap() function can be used to create heat maps, which can help to identify patterns such as associations and clusters.

Bar plots: A bar plot is a way of visualizing the relationship between a categorical variable and a numerical variable. Seaborn's barplot() function can be used to create bar plots, which can help to compare the values of a variable across different categories.

Violin plots: A violin plot is a way of visualizing the distribution of a numerical variable across different categories. Seaborn's violinplot() function can be used to create violin plots, which can help to identify patterns such as asymmetry, multi-modality, and outliers.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn also provides functions for creating line plots, box plots, swarm plots, and many other types of plots. Each plot has its own use case, and the choice of plot will depend on the data and the research question.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn.

# Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

# Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x = fmri.timepoint , y = fmri.signal )
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')
```

```python
diamonds
```

```python
sns.displot(diamonds['price'])
```

```python
sns.histplot(data=diamonds, x='price', hue='cut', kde = 'True')
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
import matplotlib.pyplot as plt

flights_pivot = flights.pivot('month', 'year', 'passengers')

sns.heatmap(data=flights_pivot, cmap='coolwarm', annot=True, fmt='d')

plt.title('Heatmap')
plt.show()
```
